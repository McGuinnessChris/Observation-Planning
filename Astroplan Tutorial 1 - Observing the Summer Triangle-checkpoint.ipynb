```python
# define observer location
from astroplan import Observer

subaru = Observer.at_site('subaru')
```

```python
# getting sky coordinates of known target stars
from astropy.coordinates import SkyCoord
from astroplan import FixedTarget

Altair = FixedTarget.from_name('Altair')
Vega = FixedTarget.from_name('Vega')

# manually entering coordinates for the third star
Deneb_coordinates = SkyCoord('20h41m25.9s', '+45d16m49.3s', frame = 'icrs')
Deneb = FixedTarget(name = 'Deneb', coord = Deneb_coordinates)
```

```python
# Want to observe at 2 a.m. Hawaii time ( Subaru is in Hawaii ), but must record UTC time
# can be done manually or by using pytz

from astropy.time import Time

time = Time('2019-06-11 12:00:00')

# or, using pytz:

import datetime as dt
import pytz

Hawaii_time = pytz.timezone('US/Hawaii').localize(dt.datetime(2019 , 6 , 11 , 2 , 0 , 0))
print(Hawaii_time)
Hawaii_time_in_UTC = Hawaii_time.astimezone( pytz.timezone('UTC') )
print(Hawaii_time_in_UTC)
```

```python
Hawaii_time = Time(Hawaii_time)
print(Hawaii_time)
```

```python
# can use Time() to remove the time difference
print(time)
print(Hawaii_time_in_UTC)
print(Time(Hawaii_time_in_UTC))

Hawaii_time_in_UTC = Time(Hawaii_time_in_UTC)
```

```python
subaru.target_is_up(Hawaii_time_in_UTC , Altair)
```

```python
subaru.target_is_up(Hawaii_time_in_UTC , Vega)
```

```python
subaru.target_is_up(Hawaii_time_in_UTC , Deneb)
```

```python
subaru.is_night(Hawaii_time_in_UTC)
```

```python
import numpy as np
import astropy.units as u

# want to observe all three in the sky at the same time
# need to find the times when each target star is above the hotizon and the sun is below the horizon

Altair_rise = subaru.target_rise_time( Hawaii_time_in_UTC , Altair ) + 5*u.minute
Altair_set = subaru.target_set_time( time , Altair ) - 5*u.minute

Vega_rise = subaru.target_rise_time( time , Vega ) + 5*u.minute
Vega_set = subaru.target_set_time( time , Vega ) - 5*u.minute

Deneb_rise = subaru.target_rise_time( time , Deneb ) + 5*u.minute
Deneb_set = subaru.target_set_time( time , Deneb ) - 5*u.minute

all_up_start = np.max( [Altair_rise , Vega_rise , Deneb_rise] )
all_up_end = np.min( [Altair_set , Vega_set , Deneb_set] )
```

```python
# finding sunset and sunrise times
sunset_tonight = subaru.sun_set_time(Hawaii_time_in_UTC , which='nearest')
sunset_tonight.iso
```

```python
sunrise_tonight = subaru.sun_rise_time(Hawaii_time_in_UTC , which='nearest')
sunrise_tonight.iso
```

```python
# finding overlapping times when the sun is down and the stars are visible
start = np.max([sunset_tonight, all_up_start])
start.iso
```

```python
end = np.min([sunrise_tonight, all_up_end])
end.iso
```

## In this part, external visibility inhibitors are examined

```python
# investigating airmass changes
from astroplan.plots import plot_airmass 
import matplotlib.pyplot as plt

plot_airmass(Altair, subaru, Hawaii_time_in_UTC) 
plot_airmass(Vega, subaru, Hawaii_time_in_UTC) 
plot_airmass(Deneb, subaru, Hawaii_time_in_UTC)  

plt.legend(loc = 'best') 
plt.show()
```

```python
# quantities produced by AltAz

Altair_altaz = subaru.altaz(Hawaii_time_in_UTC , Altair).secz
print('Altair =' , Altair_altaz)
Vega_altaz = subaru.altaz(Hawaii_time_in_UTC , Vega).secz
print('Vega =' , Vega_altaz)
Deneb_altaz = subaru.altaz(Hawaii_time_in_UTC , Deneb).secz
print('Deneb =' , Deneb_altaz)
```

```python
# plotting the parallactic angles of the stars throught the night

from astroplan.plots import plot_parallactic

plot_parallactic(Altair, subaru, Hawaii_time_in_UTC) 
plot_parallactic(Vega, subaru, Hawaii_time_in_UTC) 
plot_parallactic(Deneb, subaru, Hawaii_time_in_UTC)

plt.legend(loc = 'best') 
plt.show()

# also directly calculating the parallactic angles at our local time of 2am

print('Altair' , subaru.parallactic_angle(Hawaii_time_in_UTC , Altair))
print('Vega' , subaru.parallactic_angle(Hawaii_time_in_UTC , Vega))
print('Deneb' , subaru.parallactic_angle(Hawaii_time_in_UTC , Deneb))
```

## Accounting for the moon

```python
from astropy import units as u

# astroplan has a command to find the moon rise and set times
print('Moon rise =' , subaru.moon_rise_time(Hawaii_time_in_UTC))
print('Moon Set =' , subaru.moon_set_time(Hawaii_time_in_UTC))

# also the moon coordinates at a particular time
print('Latitude =' , subaru.moon_altaz(time).alt)
print('Longitude =' , subaru.moon_altaz(time).az)
```

```python
# start = moon rise
# end = moon set

visible_time = start + (end - start)*np.linspace(0, 1, 20)
print((subaru.moon_altaz(visible_time).alt.deg))
```

```python
# can plot these sky positions on polar plots using plot_sky

from astroplan.plots import plot_sky

Altair_style = {'color' : 'r'}
Deneb_style = {'color' : 'g'}

plot_sky(Altair, subaru, start , style_kwargs = Altair_style)  
plot_sky(Vega, subaru, start)  
plot_sky(Deneb, subaru, start, style_kwargs = Deneb_style)  

plt.legend(loc='center left', bbox_to_anchor=(1.25, 0.5))  
plt.show()  

plot_sky(Altair, subaru, end, style_kwargs=Altair_style)  
plot_sky(Vega, subaru, end)  
plot_sky(Deneb, subaru, end, style_kwargs=Deneb_style)  

plt.legend(loc='center left', bbox_to_anchor=(1.25, 0.5))  
plt.show()  
```

```python
time_window = start + (end - start) * np.linspace(0, 1, 10)

plot_sky(Altair, subaru, time_window , style_kwargs = Altair_style)  
plot_sky(Vega, subaru, time_window)  
plot_sky(Deneb, subaru, time_window, style_kwargs = Deneb_style)  

plt.legend(loc='center left', bbox_to_anchor=(1.25, 0.5))  
plt.show()  
```
