```python
# with astroplan we can define an observer by using their location
from astroplan import Observer
from astropy.coordinates import EarthLocation
import astropy.units as u
location = EarthLocation.from_geodetic(-155.4761*u.deg , 19.825*u.deg , 4139*u.m)

#location has been defined, name is defined
#timezone must be given in its .pytz format. If not known, it can be found using .pytz timexone list
subaru = Observer(location=location, name="Subaru", timezone="US/Hawaii")
print(subaru)
```

## If it is not in this list, the name and sky coordinates of the object can be specified using the following sample code for sirius

```python
from astroplan import FixedTarget
from astropy.coordinates import SkyCoord
import astropy.units as u
sirius_coord = SkyCoord(ra=101.28715533*u.deg , dec=16.71611586*u.deg)
sirius = FixedTarget(coord=sirius_coord, name="Sirius")
```

## From these coordinates, plots of the airmass and parallactic angle of the target from the observing location can be made, centred at a given time

```python
# using astroplan to make a plot of airmass centred on a given time,
# from a given observatory, looking at a given target or its coordinates
from astroplan.plots import plot_airmass
plot_airmass(target, observer, time)

# using astroplan to make a plot of parallactic angle centred on a given time,
# from a given observatory, looking at a given target or its coordinates
from astroplan.plots import plot_parallactic
plot_parallactic(target, observer, time)
```

```python
# the sample target is sirius and the sample observatory is subaru

import matplotlib.pyplot as plt
from astropy.time import Time
from astroplan.plots import plot_airmass

observe_time = Time('2019-06-12 23:30:00') # same date this code was written

plot_airmass(sirius , subaru , observe_time)
plt.show()
```

```python
# the altitude can also be displayed on the y axis

plot_airmass(sirius , subaru , observe_time , altitude_yaxis=True)
plt.show()
```

```python
# the graph can also be made to scale accordingly to the altitude instead of the airmass using plot_altitude

from astroplan.plots import plot_altitude
plt.figure(figsize = (15 , 10))
plot_altitude(sirius , subaru , observe_time, brightness_shading=True , airmass_yaxis=True)
plt.legend(shadow=True, loc=2)
```

## The time window of the plot can be changed from 24 hours to a time window of your own choice by setting a range of times
## In the example below the range is in hours from -5 to +5 of the given time

```python
import numpy as np
import matplotlib.pyplot as plt
import astropy.units as u
from astropy.time import Time
from astroplan.plots import plot_airmass

observe_time = Time('2019-06-12 23:30:00')
observe_time = observe_time + np.linspace(-5, 5, 55)*u.hour

plot_airmass(sirius , subaru , observe_time)
plt.show()
```

## Can specify the start and end times more specifically than +/- a given length of time

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.time import Time
from astroplan.plots import plot_airmass

start_time = Time('2019-06-13 20:00:00')
end_time = Time('2019-06-14 04:00:00')
delta_t = end_time - start_time
observe_time = start_time + delta_t*np.linspace(0, 1, 75)

plot_airmass(sirius , subaru , observe_time)
plt.show()
```

```python
# Or plotting altitude
plot_altitude(sirius , subaru , observe_time, airmass_yaxis=True)
```

```python
# multiple targets can be displayed on the same plot

import numpy as np
import matplotlib.pyplot as plt
from astropy.time import Time
from astroplan.plots import plot_airmass

observe_time = Time('2019-06-13 23:30:00') + np.linspace(-7.0, 5.5, 50)*u.hour

# new targets, one imported by name, one by coordinates like before

Polaris = FixedTarget.from_name('Polaris')
Deneb_coordinates = SkyCoord('20h41m25.9s', '+45d16m49.3s', frame = 'icrs')
Deneb = FixedTarget(name = 'Deneb', coord = Deneb_coordinates)

plot_airmass(sirius , subaru , observe_time)
plot_airmass(Polaris , subaru , observe_time)
plot_airmass(Deneb , subaru , observe_time)

plt.legend(shadow=True, loc=2)
plt.show()
```

```python
plot_altitude(sirius , subaru , observe_time)
plot_altitude(Polaris , subaru , observe_time)
plot_altitude(Deneb , subaru , observe_time)
plt.show()
```

```python
# Plot styles can also be changed

import matplotlib.pyplot as plt
from astroplan.plots import plot_airmass

sirius_styles = {'linestyle': '--', 'color': 'r'}
Polaris_styles = {'color': 'g'}

plot_airmass(sirius , subaru , observe_time , style_kwargs=sirius_styles)
plot_airmass(Polaris , subaru , observe_time , style_kwargs=Polaris_styles)

plt.legend(shadow=True, loc=2)
plt.show()
```

# When making observations at night, it might be beneficial to use a dark theme for any plots. For this reason, there is a dark style sheet that can be imported for use instead of the standard while of matplotlib

# Once initialized it can't be turned off

```python
from astroplan.plots import dark_style_sheet, plot_altitude
time = Time('2019-06-13 23:30:00')
plot_altitude(sirius , subaru , time, style_sheet=dark_style_sheet)
```

# You can put a visual representation of the transition from day to night on any graph by using: brightness_shading=True

```python
import matplotlib.pyplot as plt
from astropy.time import Time
from astroplan import FixedTarget, Observer
from astroplan.plots import plot_airmass

time = Time('2018-01-02 19:00')
target = FixedTarget.from_name('HD 189733')
apo = Observer.at_site('APO')

plt.figure(figsize=(15 , 5))

plt.subplot(121)
plot_airmass(target, apo, time, brightness_shading=True, altitude_yaxis=True)
plt.subplot(122)
plot_altitude(target, apo, time, brightness_shading=True, airmass_yaxis=True)
```

# Using Skyplots like in the last tutorial. Plots are centred on the zenith of the observation location. Can take a single time or a range of times to plot

```python
from astroplan.plots import plot_sky
observe_time = Time('2019-06-13 23:30:00')
plot_sky(sirius , subaru , observe_time)
plt.show()

sirius_rise = subaru.target_rise_time( observe_time , sirius ) + 5*u.minute
sirius_set = subaru.target_set_time( observe_time , sirius ) - 5*u.minute
times = sirius_rise + ( sirius_set - sirius_rise ) * np.linspace(0, 1, 10)
plot_sky(sirius , subaru , times)
plt.show()
```

```python
# can also plot multiple objects on the same plot

import matplotlib.pyplot as plt
from astropy.time import Time
from astroplan.plots import plot_sky

observe_time = Time(['2019-03-15 15:30:00'])

Vega = FixedTarget.from_name('Vega')
Deneb = FixedTarget.from_name('Deneb')
Altair = FixedTarget.from_name('Altair')

polaris_style = {'color': 'k'}
vega_style = {'color': 'g'}
deneb_style = {'color': 'r'}

plot_sky(Polaris , subaru , observe_time, style_kwargs=polaris_style)
plot_sky(Altair , subaru , observe_time)
plot_sky(Vega , subaru , observe_time , style_kwargs=vega_style)
plot_sky(Deneb , subaru , observe_time , style_kwargs=deneb_style)

plt.legend(loc='center left', bbox_to_anchor=(1.25, 0.5))
plt.show()
```

```python
# movement over time

import numpy as np
import matplotlib.pyplot as plt
import astropy.units as u
from astroplan.plots import plot_sky

observe_time = Time('2000-03-15 17:00:00')
observe_time = observe_time + np.linspace(-4, 5, 10)*u.hour

plot_sky(Altair , subaru , observe_time)

plt.legend(loc='center left', bbox_to_anchor=(1.25, 0.5))
plt.show()
```

# The north_to_east_ccw takes the planet's rotation into account. The option defaults to True, setting rotation to counter-clockwise (East on left, West on right). Setting to false changes the rotation direction to account for being in the southern hemisphere. 

```python
# astroplan uses matplotlib defaults so plots can be customized

import matplotlib.pyplot as plt
from astroplan.plots import plot_sky

guide_style = {'marker': '*'}

plot_sky(Polaris , subaru , observe_time , north_to_east_ccw=False , style_kwargs=guide_style)
plot_sky(Altair , subaru , observe_time , north_to_east_ccw=False )

plt.legend(loc='center left', bbox_to_anchor=(1.25, 0.5))
plt.show()
```

# If an observatory setup requires the zenith in the plot to be changed, the plot labels can be changed with: az_label_offset(angle)

```python
import matplotlib.pyplot as plt
from astroplan.plots import plot_sky

guide_style = {'marker': '*'}

plot_sky(Polaris , subaru , observe_time , style_kwargs=guide_style, az_label_offset=180.0*u.deg)
plot_sky(Altair , subaru ,  observe_time , az_label_offset=180.0*u.deg)

plt.legend(loc='center left', bbox_to_anchor=(1.25, 0.5))
plt.show()
```

```python
# like with a standard matplotib plot, the grid can be turned on or off

import matplotlib.pyplot as plt
from astroplan.plots import plot_sky

guide_style = {'marker': '*'}

plot_sky(Polaris , subaru , observe_time , style_kwargs=guide_style, grid=False)
plot_sky(Altair , subaru , observe_time , grid=False)

plt.legend(loc='center left', bbox_to_anchor=(1.25, 0.5))
plt.show()
```

```python
# can also alter figure size and font size like a normal matplotlib plot

# Set the figure size/font before you issue the plotting command.
plt.figure(figsize=(8,6))
plt.rc('font', size=14)

plot_sky(Polaris , subaru , observe_time)
plot_sky(Altair , subaru , observe_time)

plt.legend(loc='center left', bbox_to_anchor=(1.25, 0.5))
plt.show()

# Change font size back to default once done plotting.
plt.rc('font', size=12)
```
